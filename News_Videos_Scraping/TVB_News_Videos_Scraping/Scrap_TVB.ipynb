# TVB News Videos Scraper

## Scraping Video Metadata

```python
import json
import time
import requests
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--remote-debugging-port=9223")  


driver = webdriver.Chrome(options=chrome_options)


base_url = 'https://inews-api.tvb.com/news/entry/category?id=local&mpmLimit=0&date=20251218&lang=tc&limit=1000&page={page}&country=HK'


for page in range(1, 171):
    
    url = base_url.format(page=page)
    
    
    response = requests.get(url)
    
    if response.status_code == 200:
        
        data = response.json()
        with open(f'./JSONs/index_{page}.json', 'w', encoding='utf-8') as f:
            json.dump(data, f, ensure_ascii=False, indent=4)
        print(f"Saved index_{page}.json")
    else:
        print(f"Failed to retrieve data for page {page}")
    
    
    time.sleep(1)


driver.quit()
```

## Extracting and Cleaning Metadata

```python
import json
import os


json_dir = './JSONs'


combined_content = []


for page in range(1, 171):
    file_path = os.path.join(json_dir, f'index_{page}.json')
    
    
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
        
        
        if 'content' in data:
            combined_content.extend(data['content'])
        else:
            print(f"Warning: 'content' not found in index_{page}.json")


combined_data = {'combined_content': combined_content}


with open('./combined_content.json', 'w', encoding='utf-8') as f:
    json.dump(combined_data, f, ensure_ascii=False, indent=4)

print("All content has been combined into 'combined_content.json'")
```

```python
import json
import pandas as pd


with open('./combined_content.json', 'r', encoding='utf-8') as f:
    combined_data = json.load(f)


extracted_data = []


for item in combined_data['combined_content']:
    
    entry = {
        'id': item['id'],
        'title': item['title'].replace("\n", "").replace("\r", "").replace(",", ""),
        
        
        'url_hd': item['media']['video']['Can']['hd']['url'] if 'hd' in item['media']['video']['Can'] else None,
        'publish_datetime': item['publish_datetime'],
        'tags': ', '.join(item['tags']),
        'category_ids': ', '.join([cat['id'] for cat in item['category']])
    }
    
    
    extracted_data.append(entry)


df = pd.DataFrame(extracted_data)
```

# Downlooading Videos via FFmpeg 

```python
import requests
import subprocess
import pandas as pd


def download_m3u8_ffmpeg(m3u8_url, output_file):
    try:
        
        subprocess.run(
            ['ffmpeg', '-i', m3u8_url, '-c', 'copy', '-bsf:a', 'aac_adtstoasc', 
             "./TVB Scraped Videos/" + output_file],
            check=True
        )
        
        return True
    except subprocess.CalledProcessError as e:
        print(f"Error downloading {m3u8_url}: {e}")
        return False





def process_and_download(df):
    for index, row in df.iterrows():
        
        modified_url = row['url_hd'] + "?profile=ios"
        
        
        try:
            response = requests.get(modified_url)
            if response.status_code == 200:
                data = response.json()
                
                
                hd_url = data['content']['url']['hd']
                
                
                output_file = f"{row['id']}.mp4"
                
                
                success = download_m3u8_ffmpeg(hd_url, output_file)
                if not success:
                    print(f"Failed to download video for row {index}.")
            else:
                print(f"Failed to retrieve data for row {index}: Status Code {response.status_code}")
        except requests.exceptions.RequestException as e:
            print(f"Error requesting URL for row {index}: {e}")


process_and_download(df)
```

## Downloading Videos by using Multi-Threading

```python
import os
import requests
import subprocess
import pandas as pd
from concurrent.futures import ThreadPoolExecutor
from tqdm import tqdm


SAVE_DIR = "./TVB Scraped Videos/"


def download_m3u8_ffmpeg(m3u8_url, output_file):
    try:
        
        subprocess.run(
            ['ffmpeg', '-i', m3u8_url, '-c', 'copy', '-bsf:a', 'aac_adtstoasc', 
             os.path.join(SAVE_DIR, output_file)],
            check=True
        )
        return True
    except subprocess.CalledProcessError:
        return False


def process_row(row, progress_bar):
    
    output_file = f"{row['id']}.mp4"
    file_path = os.path.join(SAVE_DIR, output_file)
    
    
    if os.path.exists(file_path):
        progress_bar.update(1)  
        return
    
    
    modified_url = row['url_hd'] + "?profile=ios"
    
    try:
        
        response = requests.get(modified_url)
        if response.status_code == 200:
            data = response.json()
            
            
            hd_url = data['content']['url']['hd']
            
            
            if not download_m3u8_ffmpeg(hd_url, output_file):
                raise Exception(f"FFmpeg failed for ID {row['id']}")
        else:
            raise Exception(f"HTTP {response.status_code} for ID {row['id']}")
        progress_bar.update(1)  
    except Exception as e:
        print(f"Error for ID {row['id']}: {e}")
        progress_bar.update(1)  


def process_and_download_parallel(df, num_threads=4):
    
    with tqdm(total=len(df), desc="Downloading videos") as progress_bar:
        with ThreadPoolExecutor(max_workers=num_threads) as executor:
            
            futures = [executor.submit(process_row, row, progress_bar) for _, row in df.iterrows()]
            for future in futures:
                future.result()  





process_and_download_parallel(df, num_threads=8)
```
