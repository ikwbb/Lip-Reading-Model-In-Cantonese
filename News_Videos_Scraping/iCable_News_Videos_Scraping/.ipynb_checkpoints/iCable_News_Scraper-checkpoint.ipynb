# Scraping Metadata of iCable News Videos

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def get_post_data(element):
    url = element.find_element(By.TAG_NAME, "a").get_attribute("href")
    source = element.find_element(By.CLASS_NAME, "post-meta").find_elements(By.TAG_NAME, "li")[0].text
    title = element.find_element(By.CLASS_NAME, "post-title").text
    description = element.find_element(By.CLASS_NAME, "custom-text-truncate").text
    return {"url": url, "source": source, "title": title, "description": description}

def scrape_blog_posts(url_base, start_page, end_page):
    driver = webdriver.Chrome()
    for page_num in range(start_page, end_page + 1):
        url = f"{url_base}/page/{page_num}"
        driver.get(url)
        elements = WebDriverWait(driver, 30).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, "cscra-blog-post"))
        )
        for element in elements:
            data.append(get_post_data(element))
    driver.quit()
    return pd.DataFrame(data)
```

```python
import pickle

def save_variable_to_pickle(variable, filename="my_variable.pkl"):
    try:
        with open(filename, "wb") as f:
            pickle.dump(variable, f)
        print(f"Variable saved to {filename}")
    except Exception as e:
        print(f"Error saving variable to pickle file: {e}")
```

```python
from selenium.common.exceptions import TimeoutException, NoSuchElementException
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options



chrome_options = Options()
prefs = {
    "profile.managed_default_content_settings.javascript": 2  
}
chrome_options.add_experimental_option("prefs", prefs)


driver = webdriver.Chrome(options=chrome_options)


wait = WebDriverWait(driver, 5)

for index, row in datapd.iterrows():
    url = row['url']
    visited = row['visited']

    if not visited:
        try:
            driver.get(url)

            try:
                
                video_element = WebDriverWait(driver, 2).until(
                    
                    
                    lambda driver: driver.find_element(By.CSS_SELECTOR, "div.post-image")
                )
                datapd.at[index, 'visited'] = True
                continue
            except TimeoutException as e:
                pass
            except Exception as e:
                print(e)

            video_element = WebDriverWait(driver, 2).until(
                
                
                lambda driver: driver.find_element(By.CSS_SELECTOR, "video")
            )
            m3u8_url = video_element.get_attribute("data-url")
            datapd.at[index, 'm3u8_url'] = m3u8_url
            datapd.at[index, 'visited'] = True
            print(f"Found m3u8 URL for {index}: {m3u8_url}")
            save_variable_to_pickle(datapd, "datapd_with_m3u8.pkl") 

        
        
        
        
        except TimeoutException as e:  
            print(f"Timeout waiting for element on {index}")
            
        except Exception as e:  
            print(f"Other error on {index}")
            



```

# Download Videos

```python
import pandas as pd
import pickle

def load_variable_from_pickle(filename="my_variable.pkl"):
    try:
        with open(filename, "rb") as f:
            loaded_variable = pickle.load(f)
        return loaded_variable
    except FileNotFoundError:
        print(f"File {filename} not found.")
        return None
    except Exception as e:
        print(f"Error loading variable from pickle file: {e}")
        return None

datapd = load_variable_from_pickle("datapd_with_m3u8_visited.pkl") 
```

```python
datapd.head(1)
```

```python
datapd['m3u8_url_cleaned'] = datapd['m3u8_url'].apply(lambda x: None if x is not None and 'playlist.m3u8' in x else x)
```

```python
datapd["downloaded"] = False
datapd['downloaded'] = datapd.index.map(lambda x: True if x <= 2879 else False)
```

```python
datapd.iloc[18806]
```

```python
import subprocess
import pandas as pd


def download_m3u8_ffmpeg(m3u8_url, output_file):
    try:
        
        subprocess.run(
            ['ffmpeg', '-i', m3u8_url, '-c', 'copy', '-bsf:a', 'aac_adtstoasc', "./icable Scraped Videos/" + output_file],
            check=True
        )
        
        return True
    except subprocess.CalledProcessError as e:
        print(f"Error downloading {m3u8_url}: {e}")
        return False


for index, row in datapd.iterrows():
    if not row['downloaded']:  
        output_file = f"{index}.mp4"  
        
        if row["m3u8_url_cleaned"] is None:
            datapd.at[index, "downloaded"] = True
            continue

        
        success = download_m3u8_ffmpeg(row['m3u8_url_cleaned'], output_file)
        datapd.at[index, 'downloaded'] = success  


datapd.to_csv('updated_download_status.csv', index=False)

print("Process completed. Updated DataFrame:")
```
